```python
import concurrent.futures
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from concurrent.futures import ThreadPoolExecutor
import time
import sys
def go_down():
    time.sleep(5)
    current_height = driver.execute_script("return document.body.scrollHeight")
    while True:
            try:
            # Scroll to the bottom of the page
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait for some time to load new content (adjust this time based on your needs)
                time.sleep(5)

        # Get the new height of the page
                new_height = driver.execute_script("return document.body.scrollHeight")

        # If the page height has not increased, we have reached the end of the page
                if new_height == current_height:
                    break

        # Update the current height
                current_height = new_height
            except:
                break
        
def click_with_js(driver, button_selector):
    try:
        # Use JavaScript to click all buttons
        driver.execute_script(f'''
            var buttons = document.querySelectorAll('{button_selector}');
            buttons.forEach(function(button) {{
                button.click();
            }});
        ''')
    except Exception as e:
        print(f"Error clicking with JavaScript: {e}")
print("scrolling finish")
def name():
    nam=driver.find_elements(By.CSS_SELECTOR, ".mb-srp__card--title")

    index=1
    name={}

    for i in nam:
        try:
            name[index]=i.text
            index+=1
        except :
            name[index]=i.text
            index+=1
    
    print("Names collected")
    return name
        
def siz(card_elements):
    count=0
#     name={}
    size={}
    for card_element in card_elements:
        count+=1
#         nam=card_element.find_element(By.CSS_SELECTOR, ".mb-srp__card--title")
#         name[count]=nam.text
        label=card_element.find_element(By.CSS_SELECTOR, ".mb-srp__card__summary--label")
        if("AREA" in label.text):
            value=card_element.find_element(By.CSS_SELECTOR,".mb-srp__card__summary--value")
            size[count]=value.text
                
        else:
            size[count]="Element not found"
            pass
    for i,j in list(size.items()):
        try:
            b=j.split()
            if(b[1]=="sqyrd"):
                a=b[0]
                c=float(a)*9
                size[i]=c
            elif(b[1]=='sqft'):
                a=b[0]
                c=float(a)
                size[i]=c
            elif(b[1]=='sqm'):
                a=b[0]
                c=float(a)*10.76
                size[i]=c
            else:
                pass
        except:
            b=j.split()
            d=""
            e=b[0].split(",")
            for i in e:
                d+=i
            
            if(b[1]=="sqyrd"):
                a=d
                c=float(a)*9
                size[i]=c
            elif(b[1]=='sqft'):
                a=d
                c=float(a)
                size[i]=c
            elif(b[1]=='sqm'):
                a=d
                c=float(a)*10.76
                size[i]=c
            else:
                pass
    print("Size collected")
    return size

def stat(card_elements):
    count=0
#     name={}
    status={}
    comp=[]
    completion={}
    for card_element in card_elements:
        count+=1
        flag=0
        a=""
        b=""
        labels=card_element.find_elements(By.CSS_SELECTOR, ".mb-srp__card__summary--label")
        values=card_element.find_elements(By.CSS_SELECTOR,".mb-srp__card__summary--value")
        for i in range(len(labels)):
            if(len(str(labels[i].text).strip()) != 0 and flag==0):
                if("STATUS" in labels[i].text):
                    flag=1
                    a+=labels[i].text
                    b+=values[i].text
                elif("UNDER CONSTRUCTION" in labels[i].text):
                    flag=1
                    a+=labels[i].text
                    b+=values[i].text
            else:
                break
        if(a=="STATUS" and flag==1):
            status[count]=b
        elif(a=="UNDER CONSTRUCTION" and flag==1):
            status[count]=a
            comp.append(b)
        else:
            status[count]="DATA NOT FOUND"
        
                
        
#         print()
#         
#     return [status,comp]
    k=0
    for i,j in status.items():
        if(j=="UNDER CONSTRUCTION"):
            completion[i]=comp[k]
            k+=1
        elif(j=="Ready to Move"):
            completion[i]="Already completed"
        else:
            completion[i]="DATA NOT FOUND"
    output=[]
    for j in range(2):
        output.append(status)
        output.append(completion)
    print("Status Collected")
    return output

def transaction(card_elements):
    count=0
#     name={}
    tran={}
    
    for card_element in card_elements:
        count+=1
        flag=0
        a=""
        b=""
        labels=card_element.find_elements(By.CSS_SELECTOR, ".mb-srp__card__summary--label")
        values=card_element.find_elements(By.CSS_SELECTOR,".mb-srp__card__summary--value")
        for i in range(len(labels)):
            if(len(str(labels[i].text).strip()) != 0 and flag==0):
                if("TRANSACTION" in labels[i].text):
                    flag=1
                    a+=labels[i].text
                    b+=values[i].text
            else:
                break
        if(a=="TRANSACTION" and flag==1):
            tran[count]=b
            
        else:
            tran[count]="DATA NOT FOUND"
        
                
        
#         print()
    print("Transaction details collected")
    return tran

def furnish(card_elements):
    count=0
#     name={}
    fur={}
    
    for card_element in card_elements:
        count+=1
        flag=0
        a=""
        b=""
        labels=card_element.find_elements(By.CSS_SELECTOR, ".mb-srp__card__summary--label")
        values=card_element.find_elements(By.CSS_SELECTOR,".mb-srp__card__summary--value")
        for i in range(len(labels)):
            if(len(str(labels[i].text).strip()) != 0 and flag==0):
                if("FURNISHING" in labels[i].text):
                    flag=1
                    a+=labels[i].text
                    b+=values[i].text
            else:
                break
        if(a=="FURNISHING" and flag==1):
            fur[count]=b
            
        else:
            fur[count]="DATA NOT FOUND"
    print("Furnishing details collected")          
    return fur

def price():
    pri=[]
    priprsqft=[]
    price_elements = driver.find_elements(By.CSS_SELECTOR,'.mb-srp__card__price')
# print(price_elements)

    for i in price_elements:
    
        price_text = i.get_attribute('textContent').strip()
        a=price_text.split()
        flagcr=0
        flaglac=0
        flagsqft=0
#     print(price_text.split())
        for i in range(len(a)):
            if(a[i]=="Cr" and flagcr==0):
                d=float(a[i-1][1:])*10000000
                pri.append(d)
                flagcr=1
                flaglac=1
                pass
        
            elif(a[i]=="Lac" and flaglac==0):
                d=float(a[i-1][1:])*100000
                pri.append(d)
                flaglac=1
                flagcr=1
                pass
            if(a[i]=="per" ):

                val=a[i-1][1:]

                vallist=val.split(",")
                d=""

                for j in vallist:
                    d=d+j

                priprsqft.append(int(d))

            if(a[i]=='sqft'):
                flagsqft=1
            else:
                if(i>=2):
                    if(flagsqft==1 and (flagcr==1 or flaglac==1)):

                        continue
                    elif(flagsqft==0 and (flagcr==1 or flaglac==1)):
                        continue
                    elif(flagsqft==0 and (flagcr==0 or flaglac==0)):

                        pri.append("No data found")
                        priprsqft.append("No data found")
                    else:

                        continue
                else:
                    continue
                
    price={}
    priceprsqft={}
    for i,j in enumerate(pri):
        price[i+1]=j
    for i,j in enumerate(priprsqft):
        priceprsqft[i+1]=j
    output=[]
    for i in range(2):
        output.append(price)
        output.append(priceprsqft)
    print("Price details collected")
    return output
        
chrome_options = webdriver.ChromeOptions()

chrome_driver_path = r"C:\Users\pmish\chromedriver-win64\chromedriver.exe"
service = Service(executable_path=r"C:\Users\pmish\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service, options=chrome_options)


url=input("Give Url :: ")
driver.get(url)

# driver.refresh();
go_down()
print("Scrolling finish")
click_with_js(driver,'.mb-srp__card__summary__action')
card_elements = driver.find_elements(By.CSS_SELECTOR,".mb-srp__card__summary__list")
print("button clicked")
with concurrent.futures.ThreadPoolExecutor() as executor:
    future=[]
    
    future_result1 = executor.submit(name)
    future.append(future_result1.result())
    future_result2= executor.submit(siz,card_elements)
    future.append(future_result2.result())
    future_result3=executor.submit(stat,card_elements)

    result1=future_result3.result()
    future.append(result1[0])
    future.append(result1[1])
    future_result4=executor.submit(transaction,card_elements)
    future.append(future_result4.result())
    future_result5=executor.submit(furnish,card_elements)
    future.append(future_result5.result())
    future_result6=executor.submit(price)
    result2=future_result6.result()
    future.append(result2[0])
    future.append(result2[1])
concurrent.futures.wait([future_result1,future_result2,future_result3,future_result4,future_result5,future_result6])

name=future[0]
size=future[1]
status=future[2]
completion=future[3]
transaction=future[4]
furnishing=future[5]
price=future[6]
priceprsqft=future[7]
# print(name)
# print()
# print(size)
# print()
# print(status)
# print()
# print(completion)
# print()
# print(furnishing)
# print()
# print(transaction)
# print()
# print(price)
# print()
# print(priceprsqft)

print("len of name ",len(name))
print("len of size ",len(size))
print("len of status ",len(status))
print("len of completion ",len(completion))
print("len of transaction ",len(transaction))
print("len of furnishing ",len(furnishing))
print("len of price ",len(price))
print("len of priceprsqft",len(priceprsqft))


ch=input("Do you want to store complete data ?(y/n) ")
if ch.lower()=='y':

    for i,j in list(size.items()):
        try:
            if("Element not found" in j):
                del name[i]
                del size[i]
                del status[i]
                del completion[i]
                del transaction[i]
                del furnishing[i]
                del price[i]
                del priceprsqft[i]
            else:
                continue
        except:
            pass
    for i,j in list(price.items()):
        try:
            if("No data found" in j):
                del name[i]
                del size[i]
                del status[i]
                del completion[i]
                del transaction[i]
                del furnishing[i]
                del price[i]
                del priceprsqft[i]
            else:
                continue
        except:
            pass
    priceprsqftc={}    
    for i,j in list(price.items()):
        try:
            priceprsqftc[i]=j/size[i]
        except:
            pass
    print("len of name 1",len(name))
    print("len of size 1",len(size))
    print("len of status 1",len(status))
    print("len of completion 1",len(completion))
    print("len of transaction 1",len(transaction))
    print("len of furnishing 1",len(furnishing))
    print("len of price 1",len(price))
    print("len of priceprsqft 1",len(priceprsqft))
    place=input("Which place or sector you were finding ? ")
    df=pd.DataFrame(list(name.items()),columns=['Serial Number','Name of Property'])

    for key,values in size.items():
        df.loc[df['Serial Number']==key,"Size_(sqft)"]=values
    for key,values in status.items():
        df.loc[df['Serial Number']==key,"Status"]=values
    for key,values in completion.items():
        df.loc[df['Serial Number']==key,"Completion"]=values
    for key,values in furnishing.items():
        df.loc[df['Serial Number']==key,"Furnishing"]=values
    for key,values in transaction.items():
        df.loc[df['Serial Number']==key,"Transaction"]=values
    for key,values in price.items():
        df.loc[df['Serial Number']==key,"Cost_in_Rupees"]=values
# for key,values in priceprsqft.items():
#     df.loc[df['Serial Number']==key,"Cost_Per_Sqft"]=values
    for key,values in priceprsqftc.items():
        df.loc[df['Serial Number']==key,"Cost_Per_Sqft_Calculated"]=values
    
    address="D:\Gurgaon\{}[preprocessed].xlsx".format(place)   
    df.to_excel(address,index=False)
#     driver.quit()
    print("Done")
    sys.exit()
    

place=input("Which place or sector you were finding ? ")
    
for i,j in list(name.items()):
    try:
        if(place.lower() not in j.lower()):
            del name[i]
            del size[i]
            del status[i]
            del completion[i]
            del transaction[i]
            del furnishing[i]
            del price[i]
            del priceprsqft[i]
        else:
            continue
    except:
        pass

for i,j in list(size.items()):
    try:
        if("Element not found" in j):
            del name[i]
            del size[i]
            del status[i]
            del completion[i]
            del transaction[i]
            del furnishing[i]
            del price[i]
            del priceprsqft[i]
        else:
            continue
    except:
        pass
for i,j in list(price.items()):
    try:
        if("No data found" in j):
            del name[i]
            del size[i]
            del status[i]
            del completion[i]
            del transaction[i]
            del furnishing[i]
            del price[i]
            del priceprsqft[i]
        else:
            continue
    except:
        pass
priceprsqftc={}    
for i,j in list(price.items()):
    try:
        priceprsqftc[i]=j/size[i]
    except:
        pass

print("len of name 1 :: ",len(name))
print("len of size 1::",len(size))
print("len of status 1::",len(status))
print("len of completion 1::",len(completion))
print("len of transaction 1::",len(transaction))
print("len of furnishing 1 ::",len(furnishing))
print("len of price 1 ::",len(price))
print("len of priceprsqft 1 ::",len(priceprsqft))    
    
df=pd.DataFrame(list(name.items()),columns=['Serial Number','Name of Property'])

for key,values in size.items():
    df.loc[df['Serial Number']==key,"Size_(sqft)"]=values
for key,values in status.items():
    df.loc[df['Serial Number']==key,"Status"]=values
for key,values in completion.items():
    df.loc[df['Serial Number']==key,"Completion"]=values
for key,values in furnishing.items():
    df.loc[df['Serial Number']==key,"Furnishing"]=values
for key,values in transaction.items():
    df.loc[df['Serial Number']==key,"Transaction"]=values
for key,values in price.items():
    df.loc[df['Serial Number']==key,"Cost_in_Rupees"]=values
# for key,values in priceprsqft.items():
#     df.loc[df['Serial Number']==key,"Cost_Per_Sqft"]=values
for key,values in priceprsqftc.items():
    df.loc[df['Serial Number']==key,"Cost_Per_Sqft_Calculated"]=values
    
address="D:\Gurgaon\{}[preprocessed].csv".format(place)   #changed here
df.to_csv(address,index=False)
# print("Done")
driver.quit()
print("Done")
```
